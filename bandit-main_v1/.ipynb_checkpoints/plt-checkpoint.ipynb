```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
zooming_regret = np.loadtxt('zooming.csv')
zooming_regret_delta = np.loadtxt('zooming_delta.csv')

ablin_regret = np.loadtxt('ABLiN.csv')
ablin_regret_delta = np.loadtxt('ABLiN_delta.csv')

bmo_zooming_regret = np.loadtxt('bmoz.csv')
bmo_zooming_regret_delta = np.loadtxt('bmoz_delta.csv')

bmo_eli_regret = np.loadtxt('bmo_eli.csv')
bmo_eli_regret_delta = np.loadtxt('bmo_eli_delta.csv')
```

```python
plt.plot(zooming_regret, label='zooming')
plt.plot(ablin_regret, label='ABLiN')
plt.plot(bmo_zooming_regret, label='BMO_Z')
plt.plot(bmo_eli_regret, label='BMO_E')
# plt.plot(bmo_eli_regret_s2, label='BMO_E_S')
plt.legend()
plt.title('traditional regret')
plt.show()
```

```python
plt.plot(zooming_regret_delta, label='zooming')
plt.plot(ablin_regret_delta, label='ABLiN')
plt.plot(bmo_zooming_regret_delta, label='BMO_Z')
plt.plot(bmo_eli_regret_delta, label='BMO_E')
# plt.plot(bmo_eli_regret_s2_delta, label='BMO_E_S')
plt.legend()
plt.title('regret')
plt.show()
```

# std

```python
zooming_regret_std = np.loadtxt('zooming_std.csv')
zooming_regret_delta_std = np.loadtxt('zooming_std_delta.csv')

ablin_regret_std = np.loadtxt('ABLiN_std.csv')
ablin_regret_delta_std = np.loadtxt('ABLiN_std_delta.csv')

bmo_zooming_regret_std = np.loadtxt('bmoz_std.csv')
bmo_zooming_regret_delta_std = np.loadtxt('bmoz_std_delta.csv')

bmo_eli_regret_std = np.loadtxt('bmo_eli_std.csv')
bmo_eli_regret_delta_std = np.loadtxt('bmo_eli_std_delta.csv')
```

```python
plt.plot(zooming_regret_delta, label='zooming')
plt.fill_between(np.array(range(len(zooming_regret_delta))), zooming_regret_delta-zooming_regret_delta_std, 
                 zooming_regret_delta+zooming_regret_delta_std, alpha=0.1)
plt.title('delta regret zooming')
plt.show()

plt.plot(ablin_regret_delta, label='ABLiN')
plt.fill_between(np.array(range(len(ablin_regret_delta))), ablin_regret_delta-ablin_regret_delta_std, 
                 ablin_regret_delta+ablin_regret_delta_std, alpha=0.1)
plt.title('delta regret ABLiN')
plt.show()

plt.plot(bmo_zooming_regret_delta, label='BMO_Z')
plt.fill_between(np.array(range(len(bmo_zooming_regret_delta))), bmo_zooming_regret_delta-bmo_zooming_regret_delta_std, 
                 bmo_zooming_regret_delta+bmo_zooming_regret_delta_std, alpha=0.1)
plt.title('delta regret BMO_Z')
plt.show()

plt.plot(bmo_eli_regret_delta, label='BMO_E')
plt.fill_between(np.array(range(len(bmo_eli_regret_delta))), bmo_eli_regret_delta-bmo_eli_regret_delta_std, 
                 bmo_eli_regret_delta+bmo_eli_regret_delta_std, alpha=0.1)
plt.title('delta regret BMO_E')
plt.show()
# plt.plot(bmo_eli_regret_s2, label='BMO_E_S')
```
