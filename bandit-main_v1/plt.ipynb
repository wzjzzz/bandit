```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
zooming_regret = np.loadtxt('zooming.csv')
zooming_regret_delta = np.loadtxt('zooming_delta.csv')

ablin_regret = np.loadtxt('ABLiN.csv')
ablin_regret_delta = np.loadtxt('ABLiN_delta.csv')

bmo_zooming_regret = np.loadtxt('bmoz.csv')
bmo_zooming_regret_delta = np.loadtxt('bmoz_delta.csv')

bmo_eli_regret = np.loadtxt('bmo_eli.csv')
bmo_eli_regret_delta = np.loadtxt('bmo_eli_delta.csv')
```

```python
zooming_m, zooming_v = np.mean(np.cumsum(zooming_regret, axis=1), axis=0), np.std(np.cumsum(zooming_regret, axis=1), axis=0) 
ablin_m, ablin_v = np.mean(np.cumsum(ablin_regret, axis=1), axis=0), np.std(np.cumsum(ablin_regret, axis=1), axis=0) 
bmoz_m, bmoz_v = np.mean(np.cumsum(bmo_zooming_regret, axis=1), axis=0), np.std(np.cumsum(bmo_zooming_regret, axis=1), axis=0) 
bmoe_m, bmoe_v = np.mean(np.cumsum(bmo_eli_regret, axis=1), axis=0), np.std(np.cumsum(bmo_eli_regret, axis=1), axis=0)

zooming_m_d, zooming_v_d = np.mean(np.cumsum(zooming_regret_delta, axis=1), axis=0), np.std(np.cumsum(zooming_regret_delta, axis=1), axis=0) 
ablin_m_d, ablin_v_d = np.mean(np.cumsum(ablin_regret_delta, axis=1), axis=0), np.std(np.cumsum(ablin_regret_delta, axis=1), axis=0) 
bmoz_m_d, bmoz_v_d = np.mean(np.cumsum(bmo_zooming_regret_delta, axis=1), axis=0), np.std(np.cumsum(bmo_zooming_regret_delta, axis=1), axis=0)  
bmoe_m_d, bmoe_v_d = np.mean(np.cumsum(bmo_eli_regret_delta, axis=1), axis=0), np.std(np.cumsum(bmo_eli_regret_delta, axis=1), axis=0) 
```

```python
plt.plot(zooming_m, label='zooming')
plt.plot(ablin_m, label='ABLiN')
plt.plot(bmoz_m, label='BMO_Z')
plt.plot(bmoe_m, label='BMO_E')
plt.legend()
plt.title('traditional regret')
plt.savefig('./goldpr_traditional_regret.png', dpi=200)
plt.show()
```

```python
plt.plot(zooming_m_d, label='zooming')
plt.plot(ablin_m_d, label='ABLiN')
plt.plot(bmoz_m_d, label='BMO_Z')
plt.plot(bmoe_m_d, label='BMO_E')
plt.legend()
plt.title('delta regret')
plt.savefig('./goldpr_delta_regret.png', dpi=200)
plt.show()
```

# std

```python
plt.plot(zooming_m, label='zooming')
plt.fill_between(np.array(range(len(zooming_m))), zooming_m-zooming_v, 
                 zooming_m+zooming_v, alpha=0.15)
plt.title('delta regret zooming')
plt.savefig('./goldpr_zooming.png', dpi=200)
plt.show()

plt.plot(ablin_m, label='ABLiN')
plt.fill_between(np.array(range(len(ablin_m))), ablin_m-ablin_v, 
                 ablin_m+ablin_v, alpha=0.15)
plt.title('delta regret ABLiN')
plt.savefig('./goldpr_ABLiN.png', dpi=200)
plt.show()

plt.plot(bmoz_m, label='BMO_Z')
plt.fill_between(np.array(range(len(bmoz_m))), bmoz_m-bmoz_v, 
                 bmoz_m+bmoz_v, alpha=0.15)
plt.title('delta regret BMO_Z')
plt.savefig('./goldpr_bmo_z.png', dpi=200)
plt.show()

plt.plot(bmoe_m, label='BMO_E')
plt.fill_between(np.array(range(len(bmoe_m))), bmoe_m-bmoe_v, 
                 bmoe_m+bmoe_v, alpha=0.15)
plt.title('delta regret BMO_E')
plt.savefig('./goldpr_bmo_e.png', dpi=200)
# plt.legend()
plt.show()
# plt.plot(bmo_eli_regret_s2, label='BMO_E_S')
```

```python
plt.plot(zooming_m_d, label='zooming')
plt.fill_between(np.array(range(len(zooming_m_d))), zooming_m_d-zooming_v_d, 
                 zooming_m_d+zooming_v_d, alpha=0.15)
plt.title('delta regret zooming')
plt.savefig('./goldpr_zooming.png', dpi=200)
plt.show()

plt.plot(ablin_m_d, label='ABLiN')
plt.fill_between(np.array(range(len(ablin_m_d))), ablin_m_d-ablin_v_d, 
                 ablin_m_d+ablin_v_d, alpha=0.15)
plt.title('delta regret ABLiN')
plt.savefig('./goldpr_ABLiN.png', dpi=200)
plt.show()

plt.plot(bmoz_m_d, label='BMO_Z')
plt.fill_between(np.array(range(len(bmoz_m_d))), bmoz_m_d-bmoz_v_d, 
                 bmoz_m_d+bmoz_v_d, alpha=0.15)
plt.title('delta regret BMO_Z')
plt.savefig('./goldpr_bmo_z.png', dpi=200)
plt.show()

plt.plot(bmoe_m_d, label='BMO_E')
plt.fill_between(np.array(range(len(bmoe_m_d))), bmoe_m_d-bmoe_v_d, 
                 bmoe_m_d+bmoe_v_d, alpha=0.15)
plt.title('delta regret BMO_E')
plt.savefig('./goldpr_bmo_e.png', dpi=200)
# plt.legend()
plt.show()
# plt.plot(bmo_eli_regret_s2, label='BMO_E_S')
```
